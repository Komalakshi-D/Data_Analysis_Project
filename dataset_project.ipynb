# Step 7: Data Visualisation

```python
# Display the data in the columns using print function
print(data.columns)

# To create a Dataframe object to hold the data in a tabular format
df = pd.DataFrame(data.head(20))
```

```python

# Generate a pie chart to visualize the distribution of values in the 'Origin' column
df['Origin'].value_counts().plot(kind='pie')

# Add a title to the pie chart indicating the variable being visualized
plt.title("Origin")

# Display the pie chart
plt.show()
```

```python
# Generate a bar chart to visualize the distribution of values in the 'Origin' column
df['Origin'].value_counts().plot(kind='bar', color='purple')

# Add a label to the x-axis indicating the variable being visualized
plt.xlabel('Origin')

# Add a label to the y-axis indicating the count of each value
plt.ylabel('count')

# Add a title to the bar chart indicating the purpose of the visualization
plt.title('Distribution of Number of Origins')

# Display the bar chart
plt.show()
```

```python
# Generate a horizontal bar chart to visualize the distribution of values in the 'Destination' column
df['Destination'].value_counts().plot(kind='barh', color='brown')

# Add a label to the x-axis indicating the count of each value
plt.xlabel('Count')

# Add a label to the y-axis indicating the variable being visualized
plt.ylabel('Destination')

# Add a title to the horizontal bar chart indicating the purpose of the visualization
plt.title('Distribution of Destination')

# Display the horizontal bar chart
plt.show()
```

```python
# Sort the DataFrame by the 'Origin' column to ensure the data is plotted in chronological order
df.sort_values(by='Origin', inplace=True)

# Plot a line chart to visualize the trend of 'Origin' over 'Destination'
plt.plot(df['Origin'], df['Destination'], color='navy', marker='o', linestyle='-')

# Add a label to the x-axis indicating the origin
plt.xlabel('Origin')

# Add a label to the y-axis indicating the distance
plt.ylabel('Destination')

# Add a title to the line chart indicating the purpose of the visualization
plt.title('Distribution of origin to destination')

# Rotate x-axis labels by 45 degrees for better readability
plt.xticks(rotation=45)

# Add grid lines to the plot for better visualization
plt.grid(True)

# Display the line chart
plt.show()
```

```python
# Generate a scatter plot to visualize the relationship between 'Origin' and 'Destination' over 'Distance'

# determine the size
plt.figure(figsize=(10,6))

# Scatter plot for origin cities
plt.scatter(df['Origin'], df['Distance'], color='blue', label='Origin')

# Scatter plot for destination cities
plt.scatter(df['Destination'], df['Distance'], color='red', label='Destination')

# Plot title 
plt.title('Distance between Origin and Destination Cities')

# Plot x-lable
plt.xlabel('City')

# Plot y-lable
plt.ylabel('Distance (km)')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add legend
plt.legend()

# Display grid
plt.grid(True)

# Adjust layout
plt.tight_layout()

# Display the scatter plot
plt.show()
```

```python
# Generate a bar plot to visualize the relationship between 'Origin' and 'Destination' over 'Distance'

# determine the size
plt.figure(figsize=(10,6))

# Bar plot for distance from origin to destination
plt.bar(df['Origin'], df['Distance'], color='blue', label='Origin to Destination')

# Bar plot for distance from destination to origin
plt.bar(df['Destination'], df['Distance'], color='red', label='Destination to Origin')

# Plot title and labels
plt.title('Distance between Origin and Destination Cities')
plt.xlabel('City')
plt.ylabel('Distance (km)')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add legend
plt.legend()

# Display grid
plt.grid(True)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
