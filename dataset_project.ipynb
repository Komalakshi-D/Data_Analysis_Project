```python
# Define a custom function to search for rows with 'Distance' greater then 300 and less than 1000
def custom_search(row):
    return row['Distance'] > 300 and row['Distance'] < 1000

# Apply the custom function to filter rows
result = data[data.apply(custom_search, axis=1)]

# Display a result using print function
print(result.head)
```

```python
# Using Group By function
# Convert 'Distance' column to float if it's not already in float format
data['Distance'] = pd.to_numeric(data['Distance'], errors='coerce')

# Group the DataFrame by 'Origin' and calculate aggregate statistics
distance_stats = data.groupby('Origin').agg({
    'Distance': ['min', 'max'],
    
})

# Display the aggregated statistics for each Origin
print("Aggregate Statistics by Origin")

# Display a result using print function
print(distance_stats)
```

# Step 7: Data Visualisation

```python
# Display the data in the columns using print function
print(data.columns)

# To create a Dataframe object to hold the data in a tabular format
df = pd.DataFrame(data.head(20))
```

```python
# Generate a pie chart to visualize the Distribution of Travel Distances

# Define distance categories

# Define the distance bins
bins = [0, 500, 1000, float('inf')]

# Define labels for each distance category
labels = ['Short (<500 km)', 'Medium (500-1000 km)', 'Long (>1000 km)']

# Bin the data based on distance categories
df['Distance Category'] = pd.cut(df['Distance'], bins=bins, labels=labels)

# Calculate the distribution of distances
distance_distribution = df['Distance Category'].value_counts()

# Plotting the pie chart

# Set the figure size
plt.figure(figsize=(8, 8))

# Plot the pie chart
plt.pie(distance_distribution, labels=distance_distribution.index, autopct='%1.1f%%', startangle=140)

# Add title to the pie chart
plt.title('Distribution of Travel Distances')

# Display the pie chart
plt.show()
```

```python
# Generate a bar chart to visualize the Top 10 Most Traveled Routes

# Sort the data by travel distance and select the top 10 routes
top_routes = df.sort_values(by='Distance', ascending=False).head(10)

# Plotting the top 10 most traveled routes

# Set the figure size
plt.figure(figsize=(10, 6))

# Plot the bar chart
plt.bar(top_routes['Origin'] + ' to ' + top_routes['Destination'], top_routes['Distance'])

# Add a label to the x-axis 
plt.xlabel('Route')

# Add a label to the y-axis 
plt.ylabel('Distance (km)')

# Add a title to the bar chart indicating the purpose of the visualization
plt.title('Top 10 Most Traveled Routes')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# Adjust layout to prevent clipping of labels
plt.tight_layout()

# Display the bar chart
plt.show()
```

```python
# Generate a horizontal bar chart to visualize the Top 5 Distances from Delhi

# Choose a specific city (Such as "Delhi") to analyze
city = "Delhi"

# Filter the data for the chosen city
city_data = data[data["Origin"] == city]

# Sort the data by distance in descending order and select the top 5 distances
top_distances = city_data.sort_values(by="Distance", ascending=False).head()

# Plot the top 5 distances from the chosen city

# Set the figure size
plt.figure(figsize=(10, 6))

# Plot the horizontal bar chart
plt.barh(top_distances["Destination"], top_distances["Distance"], color="skyblue")

# Add a label to the x-axis 
plt.xlabel("Distance (km)")

# Add a label to the y-axis 
plt.ylabel("Destination")

# Add a title to the horizontal bar chart indicating the purpose of the visualization
plt.title(f"Top 5 Distances from {city}")

# Display the horizontal bar chart
plt.show()
```

```python
# Generate a histogram to visualize the distribution of values in the 'Distance' column

# plot the histogram 
plt.hist(df['Distance'], bins=20, color='green', edgecolor='black')

# Add a label to the x-axis indicating the variable being visualized
plt.xlabel('Distance')

# Add a label to the y-axis indicating the frequency of observations
plt.ylabel('Frequency')

# Add a title to the histogram indicating the purpose of the visualization
plt.title('Distribution of Distance')

# Add grid lines to the histogram for better visualization
plt.grid(True)

# Display the histogram
plt.show()
```

```python
# Generate a bar chart to visualize the Total Distance from Each City

# Group the data by 'Origin' and calculate the total distance from each city
total_distance = data.groupby('Origin')['Distance'].sum().sort_values(ascending=False)

# Plot the bar plot

# Set the figure size
plt.figure(figsize=(11, 8))

# Plot the bar chart
total_distance.plot(kind='bar', color='blue')

# Add a label to the x-axis 
plt.xlabel('City')

# Add a label to the y-axis 
plt.ylabel('Total Distance (km)')

# Add a title to the horizontal bar chart indicating the purpose of the visualization
plt.title('Total Distance from Each City')

# Rotate x-axis labels by 45 degrees for better readability
plt.xticks(rotation=45)

# Display the bar chart
plt.show()
```

```python
# Generate a horizontal bar chart to visualize the Top 5 Distances from Delhi

# Choose a specific origin (e.g., Delhi) to analyze
origin = "Bengaluru"

# Filter the data for the chosen origin
origin_data = data[data["Origin"] == origin]

# Plot the distances from the origin to multiple destinations
plt.figure(figsize=(10,4))

# Plot the bar chart
plt.bar(origin_data['Destination'], origin_data['Distance'], color='aqua')

# Add a label to the x-axis 
plt.xlabel('Destination')

# Add a label to the y-axis 
plt.ylabel('Distance (km)')

# Add a title to the bar chart indicating the purpose of the visualization
plt.title(f'Distances from {origin}')

# Rotate x-axis labels by 45 degrees for better readability
plt.xticks(rotation=45)

# Add grid lines to the histogram for better visualization
plt.grid(True)

# Display the bar chart
plt.show()
```

```python
# Plotting a line chart for distance between different cities

# Set the figure size
plt.figure(figsize=(12, 8))

# Plotting the line chart
plt.plot(df['Origin'] + ' to ' + df['Destination'], df['Distance'], marker='o', linestyle='-')

# Add a label to the x-axis 
plt.xlabel('Route') 

# Add a label to the y-axis 
plt.ylabel('Distance (km)')  

# Add a title to the line chart indicating the purpose of the visualization
plt.title('Distance between Different Cities')

# Rotating x-axis labels for better visibility
plt.xticks(rotation=90)

# Add grid lines to the plot for better visualization
plt.grid(True) 

# Adjust the layout to prevent overlapping content
plt.tight_layout()

# Display the line chart
plt.show()
```
